```python
#### DECORATOR EXAMPLES FROM P. 78 OF PYTHON TRICKS #####

def uppercase(func):
    def wrapper():
        original_result = func()
        modified_result = original_result.upper()
        return modified_result
    return wrapper

def lowercase(func):
    def wrapper():
        original_result = func()
        modified_result = original_result.lower()
        return modified_result
    return wrapper

@uppercase
def greet1():
    return '\nthis sentence was originally all in lowercase text; it was converted to uppercase by using the @uppercase decorator'

@lowercase
def greet2():
    return 'THIS SENTENCE WAS ORIGINALLY ALL IN UPPERCASE TEXT; IT WAS CONVERTED TO LOWERCASE BY USING THE @LOWERCASE DECORATOR\n'

print(greet1())
print(greet2())
```

```python
##### DECORATOR EXAMPLE FROM P. 81 OF PYTHON TRICKS #####

def strong(func):
    def wrapper():
        return '<strong>' + func() + '</strong>'
    return wrapper

def emphasis(func):
    def wrapper():
        return '<em>' + func() + '</em>'
    return wrapper

@strong
@emphasis
def greet():
    return 'Hello!'

print(greet())
```

```python
##### EXAMPLE OF *args AND **kwargs FROM P. 88 OF PYTHON TRICKS:  THE BOOK #####

def foo(required, *args, **kwargs):
    print(required)
    if args:
        print(args)
    if kwargs:
        print(kwargs)

foo('hello')
foo('hello', 1, 2, 3)
foo('hello', 1, 2, 3, key1='value', key2=999)
```
